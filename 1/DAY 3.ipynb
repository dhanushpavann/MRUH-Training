```python
stop_words = ['ever', 'also', "let's", 'for', "you're", 'under', 'you', 'why', 'being', 'else', "there's", "she'll", 'was', 'an', 'below', 'further', "how's", 'because', 'could', "aren't", 'how', 'him', 'up', 'am', 'too', "i'll", "you'd", "you'll", 'more', "when's", 'these', 'and', "she'd", 'this', 'until', "won't", 'do', 'its', "why's", 'into', "he'd", 'been', 'where', "we've", 'therefore', "i've", "doesn't", 'it', 'they', 'before', 'own', 'about', 'the', "they'd", 'yourselves', 'so', 'from', 'can', "mustn't", 'ourselves', 'off', 'down', 'herself', 'during', "here's", 'if', 'after', 'i', "they've", "wouldn't", 'http', 'few', 'hers', 'cannot', "can't", 'having', 'her', 'once', 'myself', 'again', 'otherwise', 'out', "they're", "didn't", 'no', 'since', 'is', 'same', "they'll", 'with', 'had', 'yourself', 'very', 'com', 'when', "where's", 'while', "you've", 'www', 'himself', 'our', "it's", "weren't", 'are', 'we', 'my', 'them', 'would', "i'm", 'such', 'which', 'all', "shouldn't", 'both', 'nor', 'what', 'between', 'only', 'then', 'any', 'has', "isn't", 'r', 'in', "shan't", 'who', 'of', 'on', 'most', 'hence', "that's", "who's", "we'd", 'to', 'his', 'or', "he's", 'not', 'however', 'theirs', 'just', 'other', 'than', 'some', "he'll", 'ours', "she's", 'over', "wasn't", 'those', 'through', 'by', "don't", 'like', 'were', 'ought', "i'd", "we're", 'should', "what's", 'be', 'does', 'here', 'yours', 'above', 'at', 'get', 'as', "we'll", 'that', 'there', "hadn't", 'your', 'k', 'themselves', "hasn't", 'itself', 'have', 'she', 'against', 'each', "couldn't", 'he', 'whom', 'but', 'doing', 'a', 'did', 'me', 'their', "haven't", 'shall']
```

```python
def stop_word_removal(text):
    tokens = text.split(' ')
    words_without_stopwords = [x for x in tokens if x not in stop_words]
    final_sentence = ' '.join(words_without_stopwords)
    return final_sentence
```

```python
df
```

```python
df['clean_data'] = df['data'].apply(stop_word_removal)
```

```python
df
```

### Get the vector embeddings

```python
def get_embeddings(text):
    doc = nlp(text)
    return doc.vector
```

```python
df["embeddings"] = df["clean_data"].apply(get_embeddings)
```

## Day 3

### Encode the classes

```python
classes = df['labels'].unique()
```

```python
classes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(classes)
```

```python
df['Encoded_labels'] = le.transform(df['labels'])
```

```python
df.tail()
```

```python
pd.set_option('display.max_colwidth',300)
```

```python
df[['Encoded_labels','labels']].drop_duplicates()
```

### Get the training data and testing data 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


X = np.vstack(df["embeddings"].values)
y = df["Encoded_labels"]

```

### Choose any Classifier model 

```python
# Create a model object
model = RandomForestClassifier(random_state=42)

```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)
```

```python

# Print the cross-validation scores
print(scores)
print(f"Mean CV accuracy: {np.mean(scores)}")
```

### Choosing a different classifier model 

```python
from sklearn.tree import DecisionTreeClassifier  
model2= DecisionTreeClassifier(criterion='gini', random_state=42)  
```

```python
scores = cross_val_score(model2, X, y, cv=5)
print(scores)
print(f"Mean CV accuracy: {np.mean(scores)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=11,)
```

```python
scores = cross_val_score(model3, X, y, cv=5)
print(scores)
print(f"Mean CV accuracy: {np.mean(scores)}")
```
