```python
import pandas as pd
df=pd.read_csv("Age,Salary,Purchase.txt")
print(df.head(6))
# Check for any non-numeric values in the 'Age' column
print(df['Age'].dtype)

```

```python
import  matplotlib.pyplot as plt

plt.figure(figsize=(5,4))
plt.scatter(df['Age'],df["Purchase"],cmap='bwr',label="Purchased")
plt.xlabel("Age")
plt.ylabel("Purchased(No=0,Yes=1)")
plt.title("Project")
plt.legend(["No Purchase", "Purchase"], loc='lower left')
plt.grid(True)
plt.show()

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split



x = df[['Age']]
y = df[['Purchase']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LogisticRegression()

model.fit(x_train, y_train)

new_age = [[59]]
pred = model.predict(new_age)
prob = model.predict_proba(new_age)

print(f"Prediction for Age {new_age[0][0]}: {'will purchase' if pred[0] == 1 else 'will not purchase'}")
```
