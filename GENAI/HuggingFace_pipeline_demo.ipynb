```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", framework="tf")
```

```python
result = sentiment_pipeline(["I love my college!",'I hate people opinion about AI', "I really liked the movie","I dont want to go outside"])
print(result)
```

# **HOW TO CHANGE FROM DEFAULT MODEL**

```python
sentiment_pipeline2 = pipeline("sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",tokenizer = "distilbert-base-uncased-finetuned-sst-2-english", framework="tf")
```

```python
result2 = sentiment_pipeline2(["I love my college!",'I hate people opinion about AI', "I really liked the movie","I dont want to go outside"])
print(result2)
```

```python
texts = [
    "This product is amazing!",
    "I'm really disappointed with the service.",
    "The weather is lovely today."
]
results2 = sentiment_pipeline2(texts)
print(results2)
```

```python
from collections import Counter

texts = [
    "I love this!",
    "This is terrible.",
    "Amazing experience!",
    "Not worth the money."
]
results2 = sentiment_pipeline2(texts)
sentiment_labels = [result['label'] for result in results2]

sentiment_distribution = Counter(sentiment_labels)
print(sentiment_distribution)
```

# OTHER EXAMPLES

```python
generator = pipeline("text-generation", model="gpt2")

prompt = "There once lived a king"
result = generator(prompt, truncation=True, num_return_sequences=1)

print(result)
```

```python
qa_pipeline = pipeline("question-answering")

context = """The Great Wall of China is a historic fortification that stretches
over 13,000 miles. It was primarily built to protect against invasions and
was constructed during the Ming Dynasty."""

question = "Who built the Great Wall of China?"

result = qa_pipeline(question=question, context=context)

print(result)
```
